## Python Operators

```python
#Arithmetic Operators

a = 14
b = 3

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a**b) #Exponentiation
print(a//b) #Floor division


#Assignment Operators
x=11
print(x)

x+=3
print(x)

x-=3
print(x)

x*=3
print(x)

x/=3
print(x)

x%=3
print(x)

x**=3
print(x)

x//=3
print(x)


#Comparison Operators
a=4
b=3

print(a==b)
print(a!=b)
print(a<b)
print(a>b)
print(a<=b)
print(a>=b)


#Logical Operators
a=23
b=20

print(a>3 and b<25)         #Returns True if both statements are true
print(a>3 or b<25)          #Returns True if one of the statements is true
print(not(a>3 and b<25))    #Reverse the result, returns False if the result is true


#Identity Operators
""" Identity operators are used to compare the objects,
not if they are equal, but if they are actually the same 
object, with the same memory location:"""

a=["Dhaka","Dhoha"]
b=["Dhaka","Dhoha"]
c=a

print(a is c)
print(a is b)
print(a == b)


print(a is not c)
print(a is not b)
print(a != b)


#Membership Operators
""" 
Membership operators are used to test if a 
sequence is presented in an object."""

x=["Dhaka","Dhoha"]
print("Dhoha" in x)     #Returns True if a sequence with the specified value is present in the object. 
print("Dubai" not in x) #Returns True if a sequence with the specified value is not present in the object.


#Bitwise Operators
""" 
Bitwise operators are used to compare (binary) numbers."""

a = 60            # 60 = 0011 1100 
b = 13            # 13 = 0000 1101 
c = 0

c = a & b;        # output:12 = 0000 1100
print(c)

c = a | b;
print(c)

c = a ^ b;
print(c)

c =~a;
print(c)

c = a << b;
print(c)

c = a >> b;
print(c)
```

## Python Collections (Arrays)


```python
# There are four collection data types in the Python programming language:
        #List
        #Tuple
        #Set
        #Dictionary
        
        
#List
"""
A list is a collection which is ordered and changeable.
In Python lists are written with square brackets[].
"""

students=["Azhar","Rifat","Majadul"]
print(students)

#Access Items
print(students[2])
print(students[1])

#Change Item Value
students[1]="Anamul"
print(students)

friends = ["Azhar", "Mamun", "Hridoy"]
for x in friends:
  print(x)

#List Length
print(len(friends))


#Add Items
"""
Using the append() method to append an item:
"""
friends.append("Kaisar")
print(friends)

friends.insert(2,"Umor")
print(friends)

#Remove Item
friends.remove("Kaisar")
print(friends)

friends.pop()
print(friends)


del  friends[1]  #The del keyword removes the specified index.
print(friends)

del  friends


#Make a copy of a list with the copy() method.
friends = ["Azhar", "Mamun", "Hridoy"]
myfriends = friends.copy()
print(myfriends)


friends.sort()
print(friends)


#Tuple
'''.......................
A tuple is a collection which is ordered and unchangeable.
In Python tuples are written with round brackets.'''

Teachers=("Rechard","Sabbir","Habib","Biplop")
print(Teachers)
Teachers=("Rechard","Sabbir","Habib","Biplop")

print(len(Teachers))

for i in Teachers:
    print(i)
    
    

#Set
'''...................
A set is a collection which is unordered and unindexed. 
In Python sets are written with curly brackets.'''


cars = {"Toyota","Volvo","BMW"}
print(cars)

#Access Items
for i in cars:
    print(i)
       
#Add Items

cars.add("Corolla")
print(cars)

cars.update(["Acura","Audi","Alfa Romeo"])
print(cars)

print(len(cars))



#Python Dictionaries
'''......................
A dictionary is a collection which is unordered, changeable and indexed. 
In Python dictionaries are written with curly brackets, and they have keys and values.'''

myinfo = {
    "name" : "Azhar",
    "Department" : "CSE",
    "Year" : "Final"
    
}
print(myinfo)

x = myinfo["Year"]
print(x)

myinfo["color"]="red"
print(myinfo)



```

## Python If ... Else


```python
age1 = 5
age2 = 20

if age2 > age1:
  print("age2 is greater than age1")


#Elif
'''
The elif keyword is pythons way of saying "if the previous
conditions were not true, then try this condition".'''

a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")



#Eles
'''
The else keyword catches anything which isn't caught by the preceding conditions.'''

x = 200
y = 33
if y > x:
  print("y is greater than x")
elif a == b:
  print("x and y are equal")
else:
  print("x is greater than y")


a = 200
b = 33

if a > b: print("a is greater than b")    #Short Hand If ... Else
    
    
    
#And
'''The and keyword is a logical operator, and is used to combine conditional statements.'''
if a > b and c > a:
  print("Both conditions are True")



#Or
'''The or keyword is a logical operator, and is used to combine conditional statements.'''

if a > b or a > c:
  print("At least one of the conditions is True")


    

```

## While Loops

```python
i = 1
while i < 10:
  print(i)
  i = i + 1
    
    
#Use of break statement
i = 1
while i < 10:
  print(i)
  if i == 5:
    break
  i = i+1



#Use of continue statement 

i = 1
while i < 10:
  i += 1 
  if i == 5:
    continue
  print(i)
```
